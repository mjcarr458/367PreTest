```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy
from IPython.display import Markdown as write

with open("grades.csv") as csv_file:
    reader = csv.reader(csv_file,delimiter=",")
    data = pd.read_csv("grades.csv")
    
```

# Data Overview

```python
print("Number of Entries: ", len(data))
print("Mean Grade", data["Grade"].mean())
print("")
```

## Histogram Grade Distribution


```python
hist_plot = data["Grade"].hist()
hist_plot.set_xlabel("Grade")
hist_plot.set_ylabel("Number of Students")
```

```python
#Seperating Data Frame by Characteristic
blue = data.query('Characteristic == "Blue" ')
red = data.query('Characteristic == "Red" ')
orange = data.query('Characteristic == "Orange" ')
```

## Histogram Breakdown Per Group

```python
red_hist = red["Grade"].hist()
red_hist.set_xlabel("Grade")
red_hist.set_ylabel("Number of Students")
red_hist.set_title("Red Group")
```

```python
blue_hist = blue["Grade"].hist()
blue_hist.set_xlabel("Grade")
blue_hist.set_ylabel("Number of Students")
blue_hist.set_title("Blue Group")
```

```python
orange_hist = orange["Grade"].hist()
orange_hist.set_xlabel("Grade")
orange_hist.set_ylabel("Number of Students")
orange_hist.set_title("Orange Group")
```

## Mean Grade per Group 

```python
red_mean = red["Grade"].mean()
blue_mean = blue["Grade"].mean()
orange_mean = orange["Grade"].mean()
print("Red Mean: ", red_mean)
print("Blue Mean: ", blue_mean)
print("Orange Mean: ", orange_mean)
labels = ["Red", "Blue", "Orange"]
means = [red_mean,blue_mean,orange_mean]
fig = plt.figure()
plt.bar(labels,means)
plt.ylim(0,100)
```
